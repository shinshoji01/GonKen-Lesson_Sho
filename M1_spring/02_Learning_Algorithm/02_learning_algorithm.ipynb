# What is a machine learning algorithm?
A machine learning algorithm is an algorithm that learns patterns from data.
This learning process can be decomposed into 3 components, the Task (T), the Performance measure (P), and the Experience (E).

Although we use the term "learning", strictly speaking, 
the model is adjusting many internal parameters by means of simple mathematical processes in order to solve a given task. 
It does this by making its current answers happy for a performance measure(P) you set
This way, the model builds a unique algorithm to solve the problem that is appropriate to the data we feed it.

This also means that for complex problems, we need to have unbiased large amounts of data.

# Task (T)
There are a lot of tasks that are extremely difficult to solve with hand-designed algorithms.

# Performance Measure (P)
To evaluate the ability of a machine learning algorithm, 
we need to employ an appropriate quantitative (objective) measure.
The choice of measure is very important, as our model is trained so as to improve the measure we use.
An obvious choice for discrete outputs is accuracy, 
where we measure the proportion of correct outputs to the total number of samples.
We can also use its "opposite", the error rate, which is the proportion of incorrect outputs.

It is diffucult but extremely important to choose a measure that is appropriate to your task.
For complex problems, you may need to research and invent a novel measure (how do you evaluate the performance of a self-driving vehicle?).
Since our goal is to process data which is not included in training process, when you evaluate your model, you also need to prepare test data which is separate from the data used in the training process.

# Experience (E)
Machine learning algorithms can be categorized as **unsupervised** or **supervised**.
The difference is in how the model experiences the training data.
An unsupervised model doesn't require any target data and it learns the probability dustribution of the training dataset.
On the other hand, a supervised model requires labels or targets, which it learns to map inputs to.
The label can be formed in various way, such as number, text, name, etc...

```python
f1, f2 = 1, 2
X, y = get_iris_2d_visualization(f1, f2)
```

## Unsupervised (Clustering)
I employ kmeans-[clustering](https://www.geeksforgeeks.org/clustering-in-machine-learning/) for the example of unsupervised model.

```python
# Train the model
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

---

As I said, it doesn't require any label (y) for training.

---

```python
decision_boundary(X, y, kmeans)
```

---

It seems to work properly.

## Supervised (Support Vector Classification)
I employ [svc](https://monkeylearn.com/blog/introduction-to-support-vector-machines-svm/), which stands for support vector classification.

```python
# train the model
svm = SVC(random_state=0).fit(X, y)
```

---

unlike kmean-clustering, it requires both X and y as input data.

---

```python
decision_boundary(X, y, svm)
```
